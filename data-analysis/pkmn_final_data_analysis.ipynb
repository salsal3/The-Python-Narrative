## Stage: Pokémon TCG API

```python
### Setup for the Pokémon TCG API Stage ###

### Create var to hold the url to Pokemon TCG API
url_pkmn_io = 'https://api.pokemontcg.io/v2/cards'
print(url_pkmn_io)

### Add format specifier to the request url
json_pkmn_io = url_pkmn_io + '?format=json'
print(json_pkmn_io, '\n')

### Execute GET request and store response
requests_pkmn_io = requests.get(json_pkmn_io)
print(requests_pkmn_io)

### Get and store JSON content. Check class types to aid in troubleshooting
content_pkmn_io = requests_pkmn_io.content
print('Class type for content_pkmn_io:', type(content_pkmn_io))
# print(content_pkmn_io)

### Format data as JSON
data_pkmn_io = requests_pkmn_io.json()
print('Class type for data_pkmn_io:', type(data_pkmn_io))

### Print JSON with added indents to check data
# print(json.dumps(data_pkmn_io, indent=4))

### Test to call data from JSON. Output should be 'Ampharos'
data_pkmn_io['data'][0]['name']
```

```python
### Convert JSON content into a DataFrame series
series_pkmn_io = pd.read_json(content_pkmn_io)
series_pkmn_io = series_pkmn_io['data']
print(series_pkmn_io.head(), '\n')

### Flatten JSON 'data' section using normalize() function to convert into DataFrame
df_pkmn = json_normalize(series_pkmn_io)
df_pkmn.set_index('name', inplace=True)
print(f'Class type for df: {type(df_pkmn)}\n')

### Print columns to identify those with market price data
print(df_pkmn.columns, '\n')

### Select columns with market price data, as well as the set info for each card to differentiate between similarly-named cards
df_market = df_pkmn[df_pkmn.columns[df_pkmn.columns.str.contains('prices.*market|set.*id|name|series')]]
print(f'Class type for market_df: {type(df_market)}\n')
print(df_market.columns)

### Fill any null prices with 0
df_market = df_market.fillna(0)

### Rename columns and index
df_market.columns = ['Set ID','Set Name','Set Series','Holo','Reverse Holo','Normal','1St Edition Holo']
df_market.index.rename('Name', inplace=True)

df_market
```

```python
### 
def s04_1_pkmn_stage():
    time.sleep(1)
    chatter('The_Good_Knight: "So our friend Joey wants to find the value of his collection of Pokémon cards."')
    chatter('The_Good_Knight: "I\'ve imported the API of a reputable online marketplace for trading card games."')
    chatter('The_Good_Knight: "I\'ve also cleaned up some of the data ahead of time, just so we could get right to solving the problem at hand."')
    chatter('The_Good_Knight: "We can always come back to it later if you\'d like to see the process."')
    time.sleep(1)
    
    chatter('The_Good_Knight: "He has quite a big collection, so let\'s just check a couple of his favorites."')
    chatter('The_Good_Knight: "A Holo \'Ampharos\' from the Secret Wonders set."')
    chatter('The_Good_Knight: "A Reverse Holo \'Deoxys\' from the Emerald set."')
    chatter('The_Good_Knight: "And a 1St Edition Holo \'Dragonite\' from the Fossil set."')
    chatter('The_Good_Knight: "When you\'re done searching, just type in \'End\'."')
# s04_1_pkmn_stage()
```

```python
### Function for searching through Pokémon card dataframe
def s04_2_search_pkmn():
    finish_search = 'The_Good_Knight: "Done searching? Great!"'
    while True:
        ### User input will have the first letter of every word capitalized by title()
        ### Ask user for the card's name
        card_name = input("What's the name of the card you\'d like to search for?")
        card_name = card_name.title()
        if card_name == 'End':
            chatter(finish_search)
            break
        else:
            card_name_search = df_market[df_market.index == card_name]

        ### Ask user for the card's set
        card_set = input("What set does the card belong to?")
        card_set = card_set.title()
        if card_set == 'End':
            chatter(finish_search)
            break
        else:
            card_set_search = card_name_search.loc[card_name_search['Set Name'] == card_set]    
        
        ### Ask user for the card's name
        card_type = input("What type of card is it?")
        card_type = card_type.title()
        if card_type == 'End':
            chatter(finish_search)
            break
        else:
            user_search = card_set_search.iloc[0][card_type]
            print(f'The market price for a {card_type} {card_name} from the {card_set} set is ${user_search}.')
        continue

# s04_2_search_pkmn()
```

```python
### Find top 10 cards in each of the four types
### Holo
top_holo = df_market[['Set Name','Holo']]
top_holo = pd.DataFrame(top_holo)
top_holo = top_holo.sort_values(by='Holo', ascending=False).head(10)
print(top_holo)
print()
plot_holo = top_holo.hvplot.scatter(dynamic=False, label='Top 10 Holo', rot=90, hover_cols=['Set Name'])

### Reverse Holo
top_r_holo = df_market[['Set Name','Reverse Holo']]
top_r_holo = pd.DataFrame(top_r_holo)
top_r_holo = top_r_holo.sort_values(by='Reverse Holo', ascending=False).head(10)
print(top_r_holo)
print()
plot_r_holo = top_r_holo.hvplot.scatter(dynamic=False, label='Top 10 Reverse Holo', rot=90, hover_cols=['Set Name'])

### Normal
top_normal = df_market[['Set Name','Normal']]
top_normal = pd.DataFrame(top_normal)
top_normal = top_normal.sort_values(by='Normal', ascending=False).head(10)
print(top_normal)
print()
plot_normal = top_normal.hvplot.scatter(dynamic=False, label='Top 10 Normal', rot=90, hover_cols=['Set Name'])

### 1St Edition Holo
top_1st_ed = df_market[['Set Name','1St Edition Holo']]
top_1st_ed = pd.DataFrame(top_1st_ed)
top_1st_ed = top_1st_ed.sort_values(by='1St Edition Holo', ascending=False).head(10)
print(top_1st_ed)
plot_1st_ed = top_1st_ed.hvplot.scatter(dynamic=False, label='Top 10 1St Edition', rot=90, hover_cols=['Set Name'])
```
